# Description des datasets
Seattle’s Building Energy Benchmarking and Reporting Program (SMC 22.920) requires owners of non-residential and multifamily buildings (20,000 square feet or larger) to track energy performance and annually report to the City of Seattle. Buildings account for 33% of Seattle's core emissions. 

The benchmarking policy supports Seattle's goals to reduce energy use and greenhouse gas emissions from existing buildings.

In 2013, the City of Seattle adopted a Climate Action Plan to achieve zero net greenhouse gas (GHG) emissions by 2050. Annual benchmarking, reporting and disclosing of building performance are foundational elements of creating more market value for energy efficiency. Per Ordinance (125000), starting with 2015 energy use performance reporting, the City of Seattle will make the data for all building 20,000 SF and larger available annually.

This update to the benchmarking mandate was passed by Seattle City Council on February 29, 2016.

### b) Réorganisation du set

On constate qu'en 2016, nous avons des variables lattitude et longitude qui n'existe pas dans le set de 2015.
Par contre en 2015, nous avons une variable location. Regardons ce qu'elle comporte :

```python
print(data_joined["Location"].values)
print("")
print(data_joined["Address"].values)
```

En 2015, nous avons tout qui est regroupé dans la variable location :
 - Adresse
 - Code Postal
 - Ville
 - Etat
 - Longitude
 - Lattitude

EN 2016, toutes ces données ont été séparées dans autant de variables. Nous allons tout remettre au même format, à savoir celui de 2016 qui est plus pratique à utiliser.

```python

for location in data_joined.itertuples():
    
    #On commence par supprimer le début de l'adresse
    if location.Location is np.nan:
        
        continue
    
    else:
        
        #print(location.Location)
        str_divided = location.Location
        
        adresse = str_divided[str_divided.find("address"):str_divided.find("city")]
        adresse = list(adresse.split(":"))
        adresse = adresse[2].strip()[1:-4]
        #print(adresse)
        
        city = str_divided[str_divided.find("city"):str_divided.find("state")]
        city = list(city.split(":"))
        city = city[1].strip()[1:-4]
        #print(city)
        
        state = str_divided[str_divided.find("state"):str_divided.find("zip")]
        state = list(state.split(":"))
        state = state[1].strip()[1:-4]
        #print(state)
        
        zipcode = str_divided[str_divided.find("zip"):str_divided.find("longitude")]
        zipcode = list(zipcode.split(":"))
        zipcode = zipcode[1].strip()[1:-6]
        #print(zipcode)
        
        longitude = str_divided[str_divided.find("longitude"):str_divided.find("latitude")]
        longitude = list(longitude.split(":"))
        longitude = longitude[1].strip()[1:-4]
        #print(longitude)
        
        latitude = str_divided[str_divided.find("latitude"):len(str_divided)]
        latitude = list(latitude.split(":"))
        latitude = latitude[1].strip()[1:-2]
        #print(latitude)
        
        
        data_joined.loc[location.Index, "Address"] = adresse
        data_joined.loc[location.Index, "City"] = city
        data_joined.loc[location.Index, "State"] = state
        data_joined.loc[location.Index, "ZipCode"] = zipcode
        
        data_joined.loc[location.Index, "Longitude"] = float(longitude)
        data_joined.loc[location.Index, "Latitude"] = float(latitude)

#On peut maintenant dropper les colonnes 2015 inutiles
data_joined.drop(["Location","Zip Codes"], axis=1, inplace=True)
```

### c) Concaténation de colonnes

Les colonnes d'émissions de GES et d'intensité d'émission de GES ont été renommées en 2016. Nous allons donc concaténer les colonnes correspondantes

```python
#Fonction pour concaténer des colonnes

def concat_col(col_2015,col_2016,new_col):

    df1 = data_2015[col_2015]
    df2 = data_2016[col_2016]
    df_total = list(df1.append(df2, ignore_index=True))
    data_joined[new_col] = df_total

    liste_col = [col_2015, col_2016]
    data_joined.drop(columns=liste_col, axis=1, inplace=True)
```

##### Traitement données manquantes

###### Imputation de données pour les champs numériques

Pour les lignes en doublons :
 - S'il y a une donnée manquante sur les deux doublons, on la remplace par la valeur de l'autre doublon
 - Si les deux lignes sont en doublons, on remplace les valeurs par la moyenne de leur indicateur respectif (Juste pour les variables quantitatives)
 
Pour les lignes uniques :
 - On remplace chaque valeur Nan (pour les indicateurs numériques), par la moyenne de l'indicateur.
 
Pour les Nan non numériques uniques :
 - Pas de remplacement envisagé

```python
print("POUR RAPPEL : Le taux de remplissage avant traitement est de ", round((donnees_remplies / donnees_totales) * 100, 2), "%.")
print("")

#On récupère les coordonnées de toutes les valeurs nan du dataframe
df_nan = data_joined.loc[data_joined.isna().any(axis=1)]

#TRAITEMENT PAR LIGNES
#On loop sur le dataframe comportant au moins un Nan par ligne
for building in df_nan.itertuples():
    
    #On récupère le n° ID du building
    buildingID = building.OSEBuildingID
    #On récupère l'année du recensement du building
    data_year = building.DataYear
    
    #On boucle sur les colonnes
    for idx, val_nan in enumerate(building):
        
        #On recherche la valeur Nan dans la ligne
        if pd.isna(val_nan):
            
            
            #On récupère le nom de la colonne de la valeur Nan
            #On met idx-1 car lorsque l'on boucle sur un building, python rajoute une colonne Index qui n'existe pas dans le dataframe
            col_val_nan = data_joined.columns.get_values()[idx-1]
            
            #On teste s'il existe un doublon au building
            df_doublon = data_joined[data_joined["OSEBuildingID"] == buildingID]
            
            if df_doublon.shape[0] == 2:
                
                df_doublon.reset_index(drop=True)
                
                #On extrait les deux valeurs
                #On commence par récupérer le numéro de colonne DataYear pour tester quelle année est représentée sur la première ligne
                col_date = df_doublon.columns.get_loc("DataYear")

                if df_doublon.iat[0, col_date] == 2015:
                    val_2015 = df_doublon.iat[0,idx-1]
                    val_2016 = df_doublon.iat[1,idx-1]
                else:
                    val_2016 = df_doublon.iat[0,idx-1]
                    val_2015 = df_doublon.iat[1,idx-1]
                
                #On teste si les deux valeurs sont des Nan
                if pd.isna(val_2015) and pd.isna(val_2016):
                    
                    #Si c'est le cas, alors on remplace la valeur nulle par la médiane arrondie de la colonne
                    if isinstance(data_joined.iat[building[0],idx-1], str) == False:
                        data_joined.iloc[building[0],idx-1] = np.ceil(data_joined[col_val_nan].median())
                else:
                    
                    #Sinon on regarde de quelle année fait partie la valeur Nan, et on la remplace par l'autre année
                    if data_year == 2015:

                        data_joined.iat[building[0],idx-1] = val_2016
                    else:

                        data_joined.iat[building[0],idx-1] = val_2015
```

```python
#TRAITEMENT DES NAN RESTANTS
#On va remplacer toutes les valeurs Nan par la médiane de leur colonne respective

for colonne in data_joined:
    
    if (data_joined[colonne].dtype == np.float64 or data_joined[colonne].dtype == np.int64):

        if data_joined[colonne].isnull().sum().sum() > 0:
            
            med_col = np.ceil(data_joined[colonne].median())
            data_joined[colonne].fillna(med_col, inplace=True)

#On recalcule le taux de remplissage après traitement
donnees_totales = data_joined.shape[0] * data_joined.shape[1]
donnees_remplies = data_joined.count().sum()

print("Sur", donnees_totales, "données possibles dans le dataset, il y a", donnees_remplies, "données disponibles.")
print("")
print("Cela représente un taux de remplissage de ", round((donnees_remplies / donnees_totales) * 100, 2), "%.")

```
